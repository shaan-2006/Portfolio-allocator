```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ta  # Technical Analysis library
from cvxopt import matrix, solvers
from datetime import datetime
import pytz
from scipy.optimize import minimize

# Suppress CVXOPT output (not used in the new optimizer)
solvers.options['show_progress'] = False

##############################################################################
# SETTINGS & PARAMETERS
##############################################################################
# Nasdaq Top 50 stocks
nasdaq_top50 = [
    "AAPL", "MSFT", "AMZN", "GOOGL", "META", "TSLA", "NVDA", "PYPL", "INTC", "CMCSA",
    "NFLX", "ADP", "CSCO", "PEP", "COST", "AVGO", "TXN", "AMGN", "QCOM", "AMD",
    "INTU", "ISRG", "MU", "MDLZ", "ADI", "FISV", "GILD", "BKNG", "SBUX", "JD",
    "LRCX", "ATVI", "KHC", "ILMN", "DOCU", "WDAY", "MCHP", "ASML", "EA", "CRWD",
    "NET", "TEAM", "VRSK", "EXC", "OKTA", "XEL", "FAST", "KDP", "IDXX", "LULU"
]
stocks = nasdaq_top50

# Set dates dynamically: use current NY time as end_date and start_date 2 years earlier
ny_tz = pytz.timezone("America/New_York")
end_dt = pd.Timestamp.now(tz=ny_tz).normalize()  # current date (normalized)
start_dt = end_dt - pd.DateOffset(years=2)
start_date = start_dt.strftime('%Y-%m-%d')
end_date   = end_dt.strftime('%Y-%m-%d')

# Backtest parameters
initial_capital = 10000.0
risk_aversion = 1         # (Not used in new optimizer; kept for reference)
momentum_lookback = 6       # in months
cov_lookback_months = 3     # in months
top_n = 10                  # select top 10 stocks monthly
w_mom   = 0.5
w_tech  = 0.3
w_fund  = 0.2

# New parameter: risk free rate (annualized, e.g., 2% per annum)
risk_free_rate = 0.02

# Scaling factor for composite signals to translate into expected returns (annualized)
scaling_factor = 0.05

##############################################################################
# DATA DOWNLOAD
##############################################################################
raw_data = yf.download(stocks, start=start_date, end=end_date, group_by='ticker', auto_adjust=False)
if isinstance(raw_data.columns, pd.MultiIndex):
    data = raw_data.xs('Adj Close', axis=1, level=1)
else:
    data = raw_data[['Adj Close']]
data.sort_index(inplace=True)
# Resample daily prices to monthly using month-end (ME)
monthly_prices = data.resample('ME').last()

##############################################################################
# FUNDAMENTAL ANALYSIS
##############################################################################
def get_fundamental_data(stocks):
    fundamentals = {}
    for stock in stocks:
        try:
            info = yf.Ticker(stock).info
            pe   = info.get('trailingPE', np.nan)
            pb   = info.get('priceToBook', np.nan)
            mcap = info.get('marketCap', np.nan)
        except Exception:
            pe, pb, mcap = np.nan, np.nan, np.nan
        fundamentals[stock] = {
            'PE': -pe if pe and pe > 0 else np.nan,
            'PB': -pb if pb and pb > 0 else np.nan,
            'MCAP': np.log(mcap) if mcap and mcap > 0 else np.nan
        }
    df = pd.DataFrame(fundamentals).T
    df.fillna(df.median(), inplace=True)
    df['FundScore'] = 0.5 * df['PE'] + 0.3 * df['PB'] + 0.2 * df['MCAP']
    return df

fundamental_df = get_fundamental_data(stocks)

##############################################################################
# TECHNICAL INDICATORS
##############################################################################
def compute_technical_indicators(series):
    try:
        rsi = ta.momentum.RSIIndicator(series, window=14).rsi().iloc[-1]
        rsi_sig = (50 - rsi) / 50
    except Exception:
        rsi_sig = 0
    try:
        ema_short = series.ewm(span=12, adjust=False).mean()
        ema_long  = series.ewm(span=26, adjust=False).mean()
        ema_sig = (ema_short.iloc[-1] - ema_long.iloc[-1]) / series.iloc[-1]
    except Exception:
        ema_sig = 0
    try:
        macd_hist = ta.trend.MACD(series).macd_diff().iloc[-1]
    except Exception:
        macd_hist = 0
    return (rsi_sig + ema_sig + macd_hist) / 3.0

##############################################################################
# HELPER FUNCTIONS FOR MEAN-VARIANCE OPTIMIZATION
##############################################################################
def calculate_portfolio_risk(w, V):
    # Portfolio risk (standard deviation)
    w = np.array(w)
    return np.sqrt(np.dot(w, np.dot(V, w)))

def calculate_portfolio_return(w, r):
    # Portfolio return
    return np.sum(w * r)

def optimize_portfolio(expected_returns, cov_matrix, target_return):
    """
    Optimize portfolio weights by minimizing portfolio risk subject to:
      - sum(weights) = 1
      - Portfolio return equals target_return
      - weights >= 0 (long-only)
    """
    n = len(expected_returns)
    init_guess = np.ones(n) / n
    constraints = [
        {'type': 'eq', 'fun': lambda w: np.sum(w) - 1},
        {'type': 'eq', 'fun': lambda w: calculate_portfolio_return(w, expected_returns) - target_return}
    ]
    bounds = [(0, 1) for _ in range(n)]
    result = minimize(lambda w: calculate_portfolio_risk(w, cov_matrix), init_guess,
                      method='SLSQP', bounds=bounds, constraints=constraints, options={'disp': False})
    return result.x

##############################################################################
# BACKTESTING & MONTHLY EFFICIENT FRONTIER WITH TANGENT LINE PLOTS
##############################################################################
portfolio_value = initial_capital
portfolio_values = []
rebal_dates = []
weights_record = []

for i in range(momentum_lookback, len(monthly_prices) - 1):
    current_date = monthly_prices.index[i]
    next_date    = monthly_prices.index[i+1]
    past_date = monthly_prices.index[i - momentum_lookback]
    
    # Momentum signal: percentage change over lookback period
    mom_signal = (monthly_prices.loc[current_date] / monthly_prices.loc[past_date]) - 1

    # Technical signals:
    tech_scores = {}
    for stock in stocks:
        try:
            series = data[stock].loc[:current_date].dropna()
            tech_scores[stock] = compute_technical_indicators(series)
        except Exception:
            tech_scores[stock] = 0
    tech_scores = pd.Series(tech_scores)
    
    # Fundamental signal:
    fund_scores = fundamental_df['FundScore']
    
    # Composite score:
    composite = (w_mom * mom_signal + w_tech * tech_scores + w_fund * fund_scores).dropna()
    
    # Select top_n stocks based on composite score:
    selected_stocks = composite.nlargest(top_n).index
    # Use composite score as a proxy for expected returns (scaled) and add risk_free_rate
    expected_returns = risk_free_rate + composite.loc[selected_stocks].values * scaling_factor

    # Covariance estimation using daily returns over the past cov_lookback_months:
    cov_start = current_date - pd.DateOffset(months=cov_lookback_months)
    daily_data = data[selected_stocks].loc[cov_start:current_date].dropna()
    if daily_data.empty or len(daily_data) < 10:
        continue
    daily_returns = daily_data.pct_change().dropna()
    cov_matrix = daily_returns.cov().values

    # Choose a target return for optimization. Here we use the mean of expected_returns.
    target_return = np.mean(expected_returns)
    try:
        weights = optimize_portfolio(expected_returns, cov_matrix, target_return)
    except Exception as e:
        # In case optimization fails, fall back to equal weights
        weights = np.ones(len(expected_returns)) / len(expected_returns)
    
    full_weights = pd.Series(0.0, index=stocks)
    full_weights[selected_stocks] = weights

    # Compute next month's return:
    period_returns = (monthly_prices.loc[next_date] / monthly_prices.loc[current_date] - 1)
    period_ret = (full_weights * period_returns).sum()
    portfolio_value *= (1 + period_ret)

    rebal_dates.append(next_date)
    portfolio_values.append(portfolio_value)
    weights_record.append(full_weights)

    print(f"Rebalance Date: {current_date.date()} -> Next Date: {next_date.date()}")
    print("Selected Stocks:", list(selected_stocks))
    print("Optimized Weights:", np.round(full_weights[selected_stocks].values, 3))
    print("Period Return:", np.round(period_ret, 4))
    print("Portfolio Value:", np.round(portfolio_value, 2))
    print("-" * 50)

    # Display Pie Chart for the month
    plt.figure(figsize=(6, 6))
    full_weights[selected_stocks].plot(kind='pie', autopct='%1.1f%%')
    plt.title(f"Portfolio Weights: {next_date.date()}")
    plt.ylabel("")
    plt.tight_layout()
    plt.show()

    # Compute the efficient frontier for this rebalancing period
    frontier_returns = []
    frontier_risks = []
    target_range = np.linspace(np.min(expected_returns), np.max(expected_returns), 50)
    for tr in target_range:
        try:
            w_opt = optimize_portfolio(expected_returns, cov_matrix, tr)
            port_return = calculate_portfolio_return(w_opt, expected_returns)
            port_risk = calculate_portfolio_risk(w_opt, cov_matrix)
            frontier_returns.append(port_return)
            frontier_risks.append(port_risk)
        except Exception:
            continue

    frontier_returns = np.array(frontier_returns)
    frontier_risks = np.array(frontier_risks)
    
    # Compute Sharpe ratios for the frontier portfolios and find the tangent (max Sharpe)
    sharpe_ratios = (frontier_returns - risk_free_rate) / frontier_risks
    max_idx = np.argmax(sharpe_ratios)
    tangent_risk = frontier_risks[max_idx]
    tangent_return = frontier_returns[max_idx]
    slope = (tangent_return - risk_free_rate) / tangent_risk

    # Prepare tangent line data
    x_line = np.linspace(0, frontier_risks.max()*1.1, 100)
    y_line = risk_free_rate + slope * x_line

    plt.figure(figsize=(10, 6))
    plt.plot(frontier_risks, frontier_returns, 'b-', label='Efficient Frontier')
    plt.plot(x_line, y_line, 'k--', label='Capital Market Line (Tangent)')
    # Mark the tangent point (optimal portfolio)
    plt.plot(tangent_risk, tangent_return, 'ro', markersize=8, label='Tangent Portfolio')
    plt.axhline(y=risk_free_rate, color='gray', linestyle='--', label='Risk-Free Rate')
    plt.xlabel('Portfolio Risk (Std. Deviation)')
    plt.ylabel('Expected Portfolio Return')
    plt.title(f'Efficient Frontier with Tangent (as of {current_date.date()})')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

##############################################################################
# PERFORMANCE METRICS & FINAL GRAPH
##############################################################################
returns = pd.Series(portfolio_values).pct_change().dropna()
total_return = portfolio_values[-1] / initial_capital - 1
sharpe = (returns.mean() / returns.std()) * np.sqrt(12)
cum_returns = pd.Series(portfolio_values)
running_max = cum_returns.cummax()
drawdown = (cum_returns - running_max) / running_max
max_drawdown = drawdown.min()

print("\n========== FINAL PERFORMANCE ==========")
print(f"Total Return: {total_return:.2%}")
print(f"Annualized Sharpe Ratio: {sharpe:.2f}")
print(f"Maximum Drawdown: {max_drawdown:.2%}")

plt.figure(figsize=(12, 6))
plt.plot(rebal_dates, portfolio_values, marker='o')
plt.title("Composite Signal Portfolio Performance (Nasdaq Top 50)")
plt.xlabel("Date")
plt.ylabel("Portfolio Value ($)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
